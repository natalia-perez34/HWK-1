```python
import matplotlib.pyplot as plt
import numpy as np

# Original x value: x = np.arange(0, 10, 0.1)
x = np.arange(0, 25, 0.25)

# Original y value: y = np.sin(x)
y = np.sin(x*2)
plt.plot(x, y, marker="o")
plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.linspace(-20,20,100)

y_sin = np.sin(x)
y_cos = np.cos(x)

fig, opt = plt.subplots(2,2)
opt[0, 0].plot(x, y_sin)
opt[0, 0].set_title('Sin(x)')
opt[1,0].plot(x, x**2)
opt[1,0].set_title('x^2')

#hwk 

opt[0,1].plot(x,y_cos)
opt[0,1].set_title('Cos(x)')

opt[1,1].plot(x,x**3)
opt[1,1].set_title('x^3')

plt.show()
```

```python
import pandas as pd

# format json file
data =data = {
    'Name': ["Sonic", "Shadow", "Twilight", "Lola", "Sofia", "Miguel", "Isabella"],
    'Age': ["22", "34", "48", "60", "28", "100", "35"],
    'Hobbies': ["Reading", "Gaming", "Cooking", "Gardening", "Photography", "Traveling", "Painting"],
    'Gender': ["Non-binary", "Male", "Female", "Male", "Female", "Non-binary", "Female"]
}

# data frame using pandas
data_frame = pd.DataFrame(data)

# for visualization
data_frame.head(2)

json_output = data_frame.head(2).to_json(orient='records', indent=4)
print(json_output)

data_frame.to_json('people_info.json', orient='records', indent=4)

# iloc to select specific rows
selected_rows_iloc = data_frame.iloc[0:4]
print("\nSelected rows using iloc (all rows):\n", selected_rows_iloc)

# Use loc to select specific columns 
selected_columns_loc = data_frame.loc[:, ['Name', 'Hobbies']]
print("\nSelected columns using loc (Name and Hobbies for all rows):\n", selected_columns_loc)
```

```python
from skimage import exposure
from skimage import io
from pylab import *

image = io.imread('C:/Users/andad/Desktop/ML/HWK 1/images/perro.png')

gamma_correction_one = exposure.adjust_gamma(image, 0.3) # Changed to 0.3, the image becomes clearer
gamma_correction_two = exposure.adjust_gamma(image, 20) # Changed to 20, the image becomes darker
sigmoid_image = exposure.adjust_sigmoid(image, cutoff=0.3, gain=33) # Used the sigmoid function 

figure(0)
io.imshow(gamma_correction_one)

figure(1)
io.imshow(gamma_correction_two)

figure(2)
io.imshow(sigmoid_image)

io.show()
```
